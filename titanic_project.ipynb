```python
import pandas as pd
titanic = pd.read_csv("titanic1.csv")
#To print first 10 rows
titanic.head(10)
```

```python
#To print last 10 rows
titanic.tail(10)
```

```python
#Age greater than 25
import numpy as np
ages = titanic["age"]
print("Number of people greater than age 25 is",len(titanic[titanic.age>25]))
        
```

```python
#Names of passengers older than 35 years
adults = titanic[titanic.age>35]
print(adults[["name","age"]])
```

```python
#MEDIAN 
titanic[["age","fare"]].median()
```

```python
#mean
titanic[["age","fare"]].mean()
```

```python
#mininum value
titanic[["age","fare"]].min()
```

```python
#maximum value
titanic[["age","fare"]].max()
```

```python
#Mean of ticket fare for each sex 
titanic[["sex","fare"]].groupby("sex").mean()
```

```python
#Mean of ticket fare for each sex and cabin
titanic[["cabin","fare"]].groupby("cabin").mean()
```
